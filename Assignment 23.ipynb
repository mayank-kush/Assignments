Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:

Use: Scatter plots are used to visualize the relationship between two continuous variables. Each data point is represented as a point on the plot, making it easy to identify patterns, clusters, or correlations between the variables.
Line Plot:

Use: Line plots are suitable for visualizing data points over a continuous interval, such as time. They are commonly used for time series data, showing trends, changes, or comparisons over time.
Histogram:

Use: Histograms are used to visualize the distribution of a single continuous variable. They help you understand the data's frequency distribution, central tendencies, and skewness.
Bar Plot:

Use: Bar plots are useful for comparing categorical data or showing the distribution of a categorical variable. They can represent data as individual bars or grouped bars, making them suitable for various data analysis tasks.
Heatmap:

Use: Heatmaps are used to visualize the relationships between two categorical variables. They are particularly helpful for displaying correlation matrices or showing how variables relate to each other in a tabular dataset.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
fmri
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal,hue=fmri.event,style=fmri.region)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
ship=sns.load_dataset('titanic')
ship
```

```python
sns.boxplot(x=ship.pclass,y=ship.age)
```

```python
sns.boxplot(x=ship.pclass,y=ship.fare)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
diamonds=sns.load_dataset('diamonds')
```

```python
sns.histplot(data=diamonds, x="price", hue="cut", bins=30, kde=True)
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris,hue='species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights=sns.load_dataset('flights')
flights
```

```python
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")
sns.heatmap(data=flights_pivot, cmap="coolwarm", annot=True, fmt="d", linewidths=0.5)
```
