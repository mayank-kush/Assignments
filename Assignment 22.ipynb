Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('title')
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:


The subplot() function in Matplotlib is used to create multiple plots (subplots) within a single figure. It allows you to arrange and display multiple plots in a grid-like fashion, making it easier to visualize and compare different data sets or aspects of your data.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1, label='Line 1', color='b')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 1')
plt.legend()
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2, label='Line 2', color='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 2')
plt.legend()
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3, label='Line 3', color='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 3')
plt.legend()
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4, label='Line 4', color='m')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line 4')
plt.legend()
plt.tight_layout()
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays summary statistics such as the median, quartiles, and potential outliers in a concise and easily interpretable way. Box plots are used to visualize the spread and central tendency of data, making them useful for identifying skewness, variability, and the presence of outliers in a dataset.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.figure(figsize=(8, 6))
data_to_plot = [box1, box2]
plt.boxplot(data_to_plot, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot Example')
plt.show()
```
