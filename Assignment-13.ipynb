Q3. What is App routing in Flask? Why do we use app routes?

In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to specific functions or views within your web application. It allows you to define how different URLs should be handled by your application and what content or functionality should be displayed when a user visits a particular URL.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL for a given endpoint (view function) by taking into account the routing rules defined in your Flask application. This function is particularly useful because it abstracts the URL construction process and makes your code more maintainable. It ensures that your URLs are consistent even if you change your routing rules.
