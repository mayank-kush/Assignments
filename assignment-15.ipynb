Q1. Explain GET and POST methods.

The GET and POST methods are used for retrivin data from a web page to a python program.

GET Method:

The primary purpose of the GET method is to request data from a server. It is used to retrieve information without causing any side-effects on the server or the data.

POST method:

 The primary purpose of the POST method is to submit data to be processed to a server. It is used to send data that may modify the server's state or result in side-effects.

Q2. Why is request used in Flask?

In Flask, a popular web framework for Python, a request object is used to handle incoming HTTP requests from clients (such as web browsers) to the server. The request object allows you to access various components of the incoming request, such as form data, query parameters, headers, cookies, and more. It plays a crucial role in building dynamic web applications by allowing you to extract and process data sent by the client.
